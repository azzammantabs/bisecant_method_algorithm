```
def f(x):
  return x**3 + (4*x**2) - 10
```

```
a = 1
b = 2
TOL = 0.00001
N0 = 100 
```

```
i = 1
FA = f(a)
```

```
while (i <= N0):
  p = (a+b) / 2
  FP = f(p)
  print("{} {} {} {} {}".format(i,a,b,p,FP))
  
  if (FP==0) or (b-a)/2 < TOL:
    print("Root is {}".format(p))
    break
  
  i= i+1
  if (FA*FP > 0):
    a = p
    FA = FP
  else:
    b = p
  
  
```

**Exercise**:

Use the Bisection method to find solutions acurate $10^{-5}$ for the following problem


1.   $x- 2^{-x}$ for $0 \leq x \leq 1$
2.   $2x \cos(2x) - (x+1)^2 =0$ for $-3 \leq x \leq -2$ and $-1 \leq x \leq 0$



No.1

```
def f(x):
  return x - 2**(-x)


a = 0
b = 1
TOL = 0.00001
N0 = 100 

i = 1
FA = f(a)

while (i <= N0):
  p = (a+b) / 2
  FP = f(p)
  print("{} {} {} {} {}".format(i,a,b,p,FP))
  
  if (FP==0) or (b-a) / 2 < TOL:
    print("Root is {}".format(p))
    break
  
  i= i+1
  if (FA*FP > 0):
    a = p
    FA = FP
  else:
    b = p
  
  
```

```
import numpy as np

def f(x):
  return 2*x*np.cos(2*x) - (x+1)**2


a = -3
b = 0
TOL = 0.00001
N0 = 100

i = 1
FA = f(a)

while (i <= N0):
  p = (a+b) / 2
  FP = f(p)
  print("{} {} {} {} {}".format(i,a,b,p,FP))
  
  if (FP==0) or (b-a) / 2 < TOL:
    print("Root is {}".format(p))
    break
  
  i= i+1
  if (FA*FP > 0):
    a = p
    FA = FP
  else:
    b = p
  
  
```
